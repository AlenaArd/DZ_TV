```python
–ó–∞–¥–∞–Ω–∏–µ_3.3

–í —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç –Ω–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç—ã A –∏ B –ø–æ—Å—Ç—É–ø–∏–ª–æ —Ä–∞–≤–Ω–æ–µ ƒ∏–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∞ –Ω–∞
—Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç C —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –ø–æ—Å—Ç—É–ø–∏–ª–æ —Å—Ç–æ–ª—åƒ∏–æ –∂–µ, —Åƒ∏–æ–ª—åƒ∏–æ –Ω–∞ A –∏ B –≤–º–µ—Å—Ç–µ.
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–∞ A —Å–¥–∞—Å—Ç –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é, —Ä–∞–≤–Ω–∞ 0.8. –î–ª—è
—Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–∞ B —ç—Ç–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 0.7, –∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–∞ C - 0.9.
–°—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª –ø–µ—Ä–≤—É—é —Å–µ—Å—Å–∏—é. –ö–∞ƒ∏–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è: a) –Ω–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–µ A ?
–±) –Ω–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–µ B ? –≤) –Ω–∞ —Ñ–∞ƒ∏—É–ª—å—Ç–µ—Ç–µ C ?
–ó–∞–º–µ—á–∞–Ω–∏–µ: –¥–∞, —ç—Ç–∞ –∑–∞–¥–∞—á–∞ –Ω–µ –Ω–∞ —Ç–µ–º—É –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏ƒ∏, –Ω–æ —Ç–µ–º–∞ –≤–∞–∂–Ω–∞—è, –∏ –µ—ë
—Å—Ç–æ–∏—Ç –∏–Ω–æ–≥–¥–∞ –æ—Å–≤–µ–∂–∞—Ç—å –≤ –ø–∞–º—è—Ç–∏.
```

```python
–í–í–µ–¥–µ–º –æ–±–æ—Ö–Ω–∞—á–µ–Ω–∏—è "+" - —Å–µ—Å—Å–∏—è —Å–¥–∞–Ω–∞. –ù–∞–π–¥–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É—á–∏—Ç—Å—è –Ω–∞ –æ–¥–Ω–æ–º –∏–∑ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–æ–≤: –ê, –í, –°
ùëÉ(A)=0.25; ùëÉ(B)=0.25; ùëÉ(C)=0.5. –û–±–æ–∑–Ω–∞—á–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —Å–¥–∞–ª —Å–µ—Å—Å–∏—é –Ω–∞ –∫–∞–∂–¥–æ–º –æ—Ç–¥–µ–ª—å–Ω–æ–º —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ:
ùëÉ(+/A)=0.8; ùëÉ(+/B)=0.7; ùëÉ(+/C)=0.9.
P(+) = ùëÉ(+/A)*ùëÉ(A)+ ùëÉ(+/B*ùëÉ(B)+ ùëÉ(+/C)*ùëÉ(C)
–°–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª—ã –ë–∞–π–µ—Å–∞: ùëÉ(A/+)=ùëÉ(+/A)*ùëÉ(A)/P(+)
–æ—Å—Ç–∞–ª—å–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏ -  ùëÉ(B/+)=ùëÉ(+/B)*ùëÉ(B)/P(+) ; ùëÉ(C/+)=ùëÉ(+/C)*ùëÉ(C)/P(+) 
                         –í—Å–µ –¥–∞–Ω–Ω—ã–µ —É –Ω–∞—Å –µ—Å—Ç—å –Ω–∞–ø–∏—à–µ–º –∫–æ–¥ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è
```

```python
import numpy as np
p_fac = [0.25, 0.25, 0.5]
p_luck = [0.8, 0.7, 0.9]
p_i = np.arange(0, len(p_fac))
f = ['A', 'B', 'C']
p=0
for i in p_i:
    pk= p_fac[i] * p_luck[i]
    #print(pk)
    p += pk
print('p(+)=', p)  

for i in p_i:
    pf= p_fac[i] * p_luck[i] / p
    print('–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –æ–Ω —É—á–∏—Ç—Å—è –Ω–∞ —Ñ–∞–∫—É–ª—å—Ç–µ—Ç–µ', f[i], '=',pf)
```
